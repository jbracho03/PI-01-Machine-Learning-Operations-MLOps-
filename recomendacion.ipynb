```python
import numpy as np
import pandas as pd
from sklearn.decomposition import NMF
```

```python
# Convertir la columna userId a un tipo de datos entero
ratings['userId'] = ratings['userId'].astype(int)

# Separar datos en entrenamiento y prueba
train_data = ratings.sample(frac=0.8, random_state=42)
test_data = ratings.drop(train_data.index)

# Crear matriz de ratings de entrenamiento
R = np.zeros((train_data['userId'].nunique(), train_data['movieId'].nunique()))
for row in train_data.itertuples():
    R[row[1]-1, row[5]-1] = row[4]

# Factorización de matrices no negativa
model = NMF(n_components=30, init='random', random_state=42)
model.fit(R)

# Calcular la matriz de ratings de prueba
R_test = np.zeros((test_data['userId'].nunique(), test_data['movieId'].nunique()))
for row in test_data.itertuples():
    R_test[row[1]-1, row[5]-1] = row[4]

# Realizar predicciones
predictions = model.transform(R_test)

# Imprimir resultados
print(predictions)
```
